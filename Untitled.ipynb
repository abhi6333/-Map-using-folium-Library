```python
import pandas as pd
```

```python
df = pd.read_csv("Cities - BSNL - Sheet1.csv")
```

```python
df.head()
```

```python
from geopy.geocoders import ArcGIS
```

```python
nom = ArcGIS()
```

```python
df['cordinates']=df['City'].apply(nom.geocode)
```

```python
df.head()
```

```python
df["latitude"] = df['cordinates'].apply(lambda x: x.latitude)
df["longitude"] = df['cordinates'].apply(lambda x: x.longitude)
```

```python
df.head()
```

```python
a_list = df[['City','latitude','longitude']].values.tolist()
```

```python
bsnl_maps = folium.Map(loaction = [23.379379735000043,79.44332654800007])
fg = folium.FeatureGroup(name ='bsnl_map')
for i in a_list:
    fg.add_child(folium.Marker(location = [i[1], i[2]],popup=folium.Popup(i[0],parse_html= True,max_width='10',show = True),icon=folium.Icon(icon_size = '20',color='green')))
bsnl_maps.add_child(fg)
#bsnl_maps.save('bsnl_map1.html')
```
